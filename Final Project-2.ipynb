```python
import turtle as turtle

def draw_rectangle(length,height,color):
    turtle.speed(100)           #set speed for drawing
    turtle.fillcolor(color)     #set color, parameter defined below
    turtle.begin_fill()         #beginning of the drawing
    for i in range(2):          #repeat two times the following steps to draw a rectangle
        turtle.forward(length)
        turtle.right(90)
        turtle.forward(height)
        turtle.right(90)
    turtle.end_fill()           #end of the drawing

def draw_star(size,color):
    turtle.color(color)         #set color
    turtle.begin_fill()
    for i in range(5):          #5 sided star, so repeat the loop 5 times
        turtle.forward(size)
        turtle.right(144)       #turn 144 degrees to the right
        turtle.forward(size)
        turtle.right(144)
    turtle.end_fill()

def get_color(color):           #define color using wikipedia color sequence
    turtle.colormode(255)
    if color == "Old Glory Red":
        return 179,25,66
    if color == "Old Glory Blue":
        return 10,49,97
    if color == "White":
        return 255,255,255
    else:
        return 0,0,0
    turtle.end_fill()

def draw_flag(height):
    turtle.setup(1000,1000)            #set up the canva size
    height = float(height)             #convert to float
    hoist = height
    fly = 1.9 * hoist                  #all dimensions based on hoist
    hoist_of_canton = hoist * (7/13)
    fly_of_canton = fly * (2/5)
    stripe_height = hoist / 13
    star = stripe_height * (4/5)
    E = hoist_of_canton / 10
    G = fly_of_canton / 14
    
    (r,g,b) = get_color("White")
    draw_rectangle(fly, hoist,(r,g,b))    #call function to draw the outer rectangle
    
    (r,g,b) = get_color("Old Glory Blue") #call function to draw the inner rectangle
    draw_rectangle(fly_of_canton, hoist_of_canton,(r,g,b))
    
    turtle.penup()
    turtle.goto(0, - hoist_of_canton - stripe_height) #select the position to start drawing
    turtle.pendown()
    (r,g,b) = get_color("Old Glory Red")
    draw_rectangle(fly,stripe_height,(r,g,b))     #call function to draw the stripe under the blue rectangle
    
    turtle.penup()
    turtle.goto(0, - hoist_of_canton - stripe_height*3)
    turtle.pendown()
    (r,g,b) = get_color("Old Glory Red")
    draw_rectangle(fly,stripe_height,(r,g,b))     #call function to draw the stripe under the blue rectangle
    
    turtle.penup()
    turtle.goto(0, - hoist_of_canton - stripe_height*5)
    turtle.pendown()
    (r,g,b) = get_color("Old Glory Red")
    draw_rectangle(fly,stripe_height,(r,g,b))     #call function to draw the stripe under the blue rectangle
    
    (r,g,b) = get_color("Old Glory Red")
    turtle.penup()
    turtle.goto(fly_of_canton, 0)
    turtle.pendown()
    draw_rectangle(fly-fly_of_canton,stripe_height,(r,g,b))#call function to draw the stripe beside the blue rectangle
    
    (r,g,b) = get_color("Old Glory Red")
    turtle.penup()
    turtle.goto(fly_of_canton, - stripe_height*2)
    turtle.pendown()
    draw_rectangle(fly-fly_of_canton,stripe_height,(r,g,b))#call function to draw the stripe beside the blue rectangle
    
    (r,g,b) = get_color("Old Glory Red")
    turtle.penup()
    turtle.goto(fly_of_canton, - stripe_height*4)
    turtle.pendown()
    draw_rectangle(fly-fly_of_canton,stripe_height,(r,g,b))#call function to draw the stripe beside the blue rectangle
    
    (r,g,b) = get_color("Old Glory Red")
    turtle.penup()
    turtle.goto(fly_of_canton, - stripe_height*6)
    turtle.pendown()
    draw_rectangle(fly-fly_of_canton,stripe_height,(r,g,b))#call function to draw the stripe beside the blue rectangle
    
    for i in range(1,10,2):                    #set up a for-loop to draw stars (5 rows, 6 columns)
        for j in range(1,12,2):
            (r,g,b) = get_color("White")
            turtle.penup()
            turtle.goto(G*j,-E*i)              #every star starts from different position according to i and j
            turtle.pendown()
            draw_star(star,(r,g,b))
        
    for k in range(2,10,2):                    #set up a for-loop to draw stars (4 rows, 5 columns)
        for f in range(2,12,2):
            (r,g,b) = get_color("White")
            turtle.penup()
            turtle.goto(G*f,-E*k)
            turtle.pendown()
            draw_star(star,(r,g,b))
            
        
def main():                                    #define a main function which includes the four function above
    height = 200
    print(draw_flag(height))
        
if __name__ == "__main__":                     
    main()                                     #call main function
```
