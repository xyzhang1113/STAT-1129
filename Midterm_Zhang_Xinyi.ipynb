```python
#Question 1
list1=[2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3]
list1.sort() #sort the list in Ascending Order

frequency = {}
for item in list1:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1
        
for key, value in frequency.items():
    print(f'{key}:{value}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2,4,6,8,4,5,2,1,9,0,4,6,7,4,3,2,1,9,10,3,7,9,6,0,1,3,5,6,7,8,9,10,2,3,6,8,9,10,6,7,4,3])

plt.hist(x,bins=range(12),rwidth=0.8,color="hotpink")
plt.grid(axis='y', alpha=0.5)
plt.xlabel('Element in list1',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title('Frenquency Histogram',fontsize=20)

plt.show()
```

```python
#Question 2
import pandas as pd

df = pd.read_csv('amazon_orders.csv')
df.head()
```

```python
df.shape #DataFrame has 125 rows and 23 columns
```

```python
df = df.fillna(0) #fill in all NaN values with 0
df.head()
```

```python
#Deleting symbol $ to ''(nothing) and converting string to float
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#How much I've spent on Amazon from October 2019 to October 2021: $4459
df["Total Charged"].sum()
```

```python
#Average spend on Amazon: $36
df["Total Charged"].mean()
```

```python
#Median spend on Amazon in order to avoid outliers: $23
df["Total Charged"].median()
```

```python
#Maximum spend on Amazon: $305.88
df["Total Charged"].max()
```

```python
#Minimum spend on Amazon: $4.89
df["Total Charged"].min()
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#How much I've paid in Sales Tax: $222
df["Tax Charged"].sum()
```

```python
#Rate:4.98%
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
#How much I've spent on daily basis
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
#How much I've spent on monthly basis
monthly_orders = df.groupby(pd.Grouper(key='Order Date', freq='1M')).sum()["Total Charged"]
monthly_orders.head()
```

```python
#Visualizing my Amazon payment
daily_orders.plot.bar(figsize=(20,10),color='#E74C3C',ylabel='Total Charged',title='Total Charged Per Day')
plt.grid(axis='y', alpha=0.7)
```

```python
monthly_orders.plot.bar(figsize=(20,10),color='#2ECC71',ylabel='Total Charged',title='Total Charged Per Month')
plt.grid(axis='y', alpha=1)
```
