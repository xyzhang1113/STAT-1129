```r
#Question 2
df <- read.csv("amazon_orders.csv",
               header = TRUE)
head(df)
```

```r
dim(df) #Data frame has 125 rows and 23 columns
```

```r
df[is.na(df)] <- 0 #fill in all NA values with 0
head(df)
```

```r
#Deleting symbol $ to ''(nothing) and converting string to float
df$Total.Charged <- as.character(df$Total.Charged)
df$Total.Charged
typeof(df$Total.Charged)
```

```r
df$Total.Charged <- gsub('^.','',as.character(df$Total.Charged))
df$Total.Charged
typeof(df$Total.Charged)
```

```r
df$Total.Charged <- as.numeric(df$Total.Charged)
df$Total.Charged
```

```r
#How much I've spent on Amazon from October 2019 to October 2021: $4459
sum(df$Total.Charged)
```

```r
#Average spend on Amazon: $36
mean(df$Total.Charged)
```

```r
#Median spend on Amazon in order to avoid outliers: $23
median(df$Total.Charged)
```

```r
#Maximum spend on Amazon: $305.88
max(df$Total.Charged)
```

```r
#Minimum spend on Amazon: $4.89
min(df$Total.Charged)
```

```r
df$Tax.Charged <- as.character(df$Tax.Charged)
df$Tax.Charged
typeof(df$Tax.Charged)
```

```r
df$Tax.Charged <- gsub('^.','',as.character(df$Tax.Charged))
df$Tax.Charged
typeof(df$Tax.Charged)
```

```r
df$Tax.Charged <- as.numeric(df$Tax.Charged)
df$Tax.Charged
```

```r
#How much I've paid in Sales Tax: $222
sum(df$Tax.Charged)
```

```r
#Rate:4.98%
sum(df$Tax.Charged) / sum(df$Total.Charged)
```
