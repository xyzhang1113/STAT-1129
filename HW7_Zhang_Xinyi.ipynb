```r
#Question 1
data_frame <- data.frame(
    id = c(1,2,3,4,5),
    name = c("Peter","Amy","Ryan","Gary","Michelle"),
    salary = c(623.3,515.2,611.00,729.00,843.25)
)

data_frame
```

```r
New_col_DF <- cbind(data_frame, department = c("IT","chemistry","art","math","finance"))
New_col_DF
```

```r
New_col_DF[c(1,3,5),c(2,3)]
```

```r
barplot(New_col_DF$salary[c(1,4,5)], names.arg = New_col_DF$name[c(1,4,5)])
```

```r
x <- c(max(New_col_DF$salary),min(New_col_DF$salary),median(New_col_DF$salary))
labels <- c("Maximum","Minimum","Median")
pie(x, label = labels, main = "Salary Info")
```
